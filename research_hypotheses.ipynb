## Research hypotheses — stated before data analysis

### H1: Closed distillery premium
Bottles from closed distilleries (Port Ellen, Brora, Karuizawa, 
Rosebank, Littlemill) command a statistically significant price premium
over comparable bottles from operational distilleries, controlling for
age, vintage, and cask type. This premium should be accelerating over
time as remaining stock diminishes.

### H2: Founder-era Samaroli premium
Samaroli bottlings with bottling date pre-2017 (founder Silvano 
Samaroli still alive) command a significantly higher price than 
post-2017 Samaroli bottlings, controlling for distillery, age, 
and vintage. Hypothesis: the market has partially but not fully 
corrected for the quality change.

### H3: GlenDronach 1993 vintage effect
GlenDronach lots with distillation date in 1993 command a premium
over other GlenDronach vintages. The 19/3/1993 specific date carries
an additional premium within the 1993 cohort. The broad 1993 effect
is real; the specific date effect may be partly psychological.

### H4: Informational asymmetry in residuals
A baseline hedonic pricing model using only observable signals 
(age, distillery, ABV, volume, cask type) will show systematic
residuals for bottles requiring domain knowledge to value correctly.
Lots with large positive residuals represent potential undervaluation
by uninformed buyers.

### H5: ABV premium is distillery-dependent
Higher ABV commands a price premium in general, but this effect
is weaker or absent for distilleries/bottlers where diluted 
expressions are historically celebrated (e.g. old G&M at 40%,
certain Springbank society bottlings).

### H6: OB vs IB depends on distillery status
For operational distilleries with strong OB programs, OB commands
a premium. For closed distilleries, IB is the only source and
bottler reputation drives price instead.
