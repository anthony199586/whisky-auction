# Whisky auction analysis
## Week 1 — Python fundamentals

Working through core Python concepts using whisky auction data.
Covers: variables, lists, dicts, f-strings, list comprehensions, functions.

```python
bottle_name = "Springbank 1997 25 Year Old"
hammer_price = 1250.00
age_years = 25

print(bottle_name)
print(hammer_price)
print(age_years)
```

```python
print(f"Bottle: {bottle_name}")
print(f"Sold for: £{hammer_price:.2f}")
print(f"Age: {age_years} years old")
```

```python
lots = [
    {"name": "Springbank 1997 25yo", "distillery": "Springbank", "price": 1250.00},
    {"name": "Caol Ila 2002 18yo",   "distillery": "Caol Ila",   "price":  480.00},
    {"name": "Ardbeg 1975 30yo",     "distillery": "Ardbeg",     "price": 4200.00},
    {"name": "Laphroaig 10yo",       "distillery": "Laphroaig",  "price":   55.00},
    {"name": "Glenfarclas 1965 40yo","distillery": "Glenfarclas","price": 8800.00},
]

for lot in lots:
    print(f"{lot['name']:35s}  £{lot['price']:>8.2f}")
```

```python
expensive = [lot for lot in lots if lot["price"] > 1000]

print(f"Lots over £1,000: {len(expensive)}")
for lot in expensive:
    print(f"  {lot['name']} — £{lot['price']:,.0f}")
```

```python
def describe_lot(lot):
    age_info = f"{lot['name']}"
    return f"{age_info:35s} | £{lot['price']:>8,.0f} | {lot['distillery']}"

def total_value(lot_list):
    total = 0
    for lot in lot_list:
        total = total + lot["price"]
    return total

for lot in lots:
    print(describe_lot(lot))

print(f"\nTotal auction value: £{total_value(lots):,.0f}")
print(f"Average price:       £{total_value(lots) / len(lots):,.0f}")
```

```python
def summarise_by_distillery(lot_list):
    distilleries = {}

    for lot in lot_list:
        name = lot["distillery"]

        if name not in distilleries:
            distilleries[name] = {"count": 0, "total": 0}

        distilleries[name]["count"] += 1
        distilleries[name]["total"] += lot["price"]

    return distilleries

summary = summarise_by_distillery(lots)

print(f"{'Distillery':20s}  {'Lots':>4}  {'Avg price':>10}")
print("-" * 40)
for distillery, data in summary.items():
    avg = data["total"] / data["count"]
    print(f"{distillery:20s}  {data['count']:>4}  £{avg:>8,.0f}")
```
